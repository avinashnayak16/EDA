```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Load the data
file_path = 'train.csv'
data = pd.read_csv(file_path)

# Display column names to identify the correct target column name
print(data.columns)

# Replace 'metastatic_diagnosis_period' with the actual target column name
target_column = 'metastatic_diagnosis_period'  # Update this if the column name is different

# Separate features and target variable
X = data.drop(target_column, axis=1)
y = data[target_column]

# List of numerical and categorical columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Preprocess the data
X_preprocessed = preprocessor.fit_transform(X)

# Split the data
X_train, X_val, y_train, y_val = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Save preprocessed data to CSV files
train_data = np.hstack((X_train, y_train.values.reshape(-1, 1)))
val_data = np.hstack((X_val, y_val.values.reshape(-1, 1)))

train_df = pd.DataFrame(train_data)
val_df = pd.DataFrame(val_data)
print(val_df.columns)
train_df.to_csv('preprocessed_train.csv', index=False, header=False)
val_df.to_csv('preprocessed_val.csv', index=False, header=False)
```
